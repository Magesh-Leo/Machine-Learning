```python
data={'yyyy-mm-dd':['2000-06-27','2002-09-24','1999-11-29']}
df=pd.DataFrame(data)
df
```

```python
#APPLY FUNC()
def get_year(column):
    return column.split('-')[0]
df['year']=df['yyyy-mm-dd'].apply(get_year)
df
```

```python
def get_age(pre_year,current_year):
    return current_year-int(pre_year)
df['age']=df['year'].apply(get_age,current_year=2020)
df
```

```python
def get_introduce(age,pre,suf):
    return pre + str(age) + suf
df['introduce']=df['age'].apply(get_introduce,pre='I am ',suf=' years old')
df
```

```python
def get_intro(row):
    return "i was born in "+str(row.year)+" my age is "+str(row.age)
df['full_intro']=df.apply(get_intro,axis=1)   #here we must apply axis=1 coz default they know row of index value 
df
```
